```python
import face_recognition
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt

img_with_face = face_recognition.load_image_file("girl2.png")

face_locations = face_recognition.face_locations(img_with_face)

largest_face_location = None
largest_area = 0

for (top, right, bottom, left) in face_locations:
    area = (bottom - top) * (right - left)
    if area > largest_area:
        largest_area = area
        largest_face_location = (top, right, bottom, left)

if largest_face_location is not None:
    top, right, bottom, left = largest_face_location
    face_image = img_with_face[top:bottom, left:right]

    plt.imshow(face_image)
    face_pil_image = Image.fromarray(face_image)

    draw = ImageDraw.Draw(face_pil_image)
    draw.line([(0, 0), face_pil_image.size], fill="red", width=5)
    draw.line([(0, face_pil_image.size[1]), (face_pil_image.size[0], 0)], fill="red", width=5)

    face_pil_image.save("face_cutout.png")
    plt.imshow(face_pil_image)
    plt.show()
else:
    print("Không tìm thấy.")
```
